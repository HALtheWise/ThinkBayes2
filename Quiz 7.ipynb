```python
import numpy as np
import pymc3 as pm

import thinkbayes2
from thinkbayes2 import Pmf, Cdf, Suite, Joint, MakeJoint, MakeMixture
import thinkplot
```

Evaluate this PDF for x = 2, k = 3, and  = 2 using NumPy functions and scipy.special.gamma.

```python
from scipy.special import gamma
```

```python
def pdf(x, k, theta):
    return 1 / (gamma(k) * theta ** k) * x ** (k-1) * np.exp(-x / theta)
```

```python
pdf(2, 3, 2)
```

```python
# Plot distribution for k=3, theta=2
pmf = Pmf()
for x in np.linspace(0, 20):
    pmf[x] = pdf(x, 3, 2)
pmf.Normalize()

thinkplot.Pdf(pmf)
```

During the last three days in the Boston area, we have measured the following rainfalls in
inches: 0.78, 0.87, 0.64.

Use this data to compute a joint posterior distribution for the parameters of the gamma
distribution.

You can use the following priors:

- For k, Half-normal with  = 0.5.
- For , Half-normal with  = 4.

What are the posterior means for k and ?

```python
class Rainfall(Suite, Joint):
    def Likelihood(self, data, hypo):
        k, theta = hypo
        
        return pdf(data, k, theta)
```

```python
k_prior = thinkbayes2.MakeNormalPmf(mu=0, sigma=0.5, num_sigmas=3, n=201)
k_prior = Pmf({k: p for k, p in k_prior.Items() if k > 0})

theta_prior = thinkbayes2.MakeNormalPmf(mu=0, sigma=4, num_sigmas=3, n=201)
theta_prior = Pmf({t: p for t, p in theta_prior.Items() if t > 0})

prior = Rainfall(MakeJoint(k_prior, theta_prior))
```

```python
posterior = prior.Copy()

posterior.UpdateSet([0.78, 0.87, 0.64])
```

```python
k_post = posterior.Marginal(0)
theta_post = posterior.Marginal(1)
```

```python
thinkplot.Pdf(k_prior, label='prior', color='grey')
thinkplot.Pdf(posterior.Marginal(0), label='posterior')
thinkplot.Config(title='K')

posterior.Marginal(0).Mean()
```

```python
thinkplot.Pdf(theta_prior, label='prior', color='grey')
thinkplot.Pdf(posterior.Marginal(1), label='posterior')
thinkplot.Config(title='Theta')

posterior.Marginal(1).Mean()
```

Generate a predictive distribution for the amount of rain we will get tomorrow (assuming
that it rains at all). What is the predictive mean?

```python
metapmf = Pmf()

xs = np.linspace(0, 6)[1:]

for (k, theta), p in posterior.Items():
    metapmf[Pmf({x:pdf(x, k, theta) for x in xs})] = p
    
mixture = MakeMixture(metapmf)
```

```python
thinkplot.Pdf(mixture)
mixture.Mean()
```
